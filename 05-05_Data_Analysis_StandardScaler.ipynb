```python
# Code below is for visualizing the dataset 

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Loading the data
df = pd.read_csv('winequality-red.csv', header = 0, delimiter=';')
headers = df.columns
data = df.values

# Using all features for PCA visualization
X = df.drop(['quality'],axis=1)
y = df.quality

# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Applying PCA 
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Visualizing the data to a reduced dimensional space
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.colorbar(label="Class")
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA Visualization')
plt.show()
```

```python
# Obtaining classes 5 & 6 for PCA visualization
df_5_6 = df.loc[df['quality'].isin([5,6])]
X = df_5_6.drop(['quality'],axis=1)
y = df_5_6.quality

# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Applying PCA 
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Visualizing the data to a reduced dimensional space
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.colorbar(label="Class")
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA Visualization Classes 5 & 6')
plt.show()
```
